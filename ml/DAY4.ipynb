```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import os

# The 'FileNotFoundError' indicates that the file was not accessible in Colab.
# Please upload the 'Crop_recommendation.csv' file to your Colab session.
# You can use the following code to upload the file:

# Check if the file already exists to avoid re-uploading
file_name = "Crop_recommendation.csv"
if not os.path.exists(file_name):
    uploaded = files.upload()
    for fn in uploaded.keys():
        print(f'User uploaded file "{fn}"')

# Now try to read the CSV file
data = pd.read_csv(file_name)
print(data.head())
```

```python
import plotly.express as px
fig = px.pie(data,names="label",title="Crop Distribution")
fig.update_layout(
    width=700,
    height=700)
fig.show()
```

```python
for col in data.columns:
    plt.figure(figsize=(10,4))
    sns.histplot(data[col],kde =(col!='label'),color ='red')
    plt.title(col)
    if col == 'label':
        plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
df1 = data.copy()

le = LabelEncoder()

for col in df1.select_dtypes(include='object').columns:
    df1[col] = le.fit_transform(df1[col].astype(str))

corr_matrix = df1.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
